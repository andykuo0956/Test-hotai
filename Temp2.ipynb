```python
import pandas as pd
from keplergl import KeplerGl
import geopandas as gpd
```

```python
df_yoxi = pd.read_csv('/Users/guoyanting/Desktop/NTUDAC/比賽/和泰/rawData/和泰MaaS/yoxi.csv',engine='python',encoding='utf-8')
```

```python
df_yoxi
```

```python
df_yoxi_1 = df_yoxi
```

```python
df_yoxi_1['上車經緯度_1'] = df_yoxi['上車經緯度'].str.replace("POINT","")
```

```python
df_yoxi_1['下車經緯度_1'] = df_yoxi['下車經緯度'].str.replace("POINT","")
```

```python
df_yoxi_1['下車經緯度_2'] = df_yoxi_1['下車經緯度_1'].str.replace("(","")
df_yoxi_1['下車經緯度_2'] = df_yoxi_1['下車經緯度_2'].str.replace(")","")
df_yoxi_1['上車經緯度_2'] = df_yoxi_1['上車經緯度_1'].str.replace("(","")
df_yoxi_1['上車經緯度_2'] = df_yoxi_1['上車經緯度_2'].str.replace(")","")
```

```python
df_yoxi_1[['Why','Why2']] = df_yoxi_1['上車經緯度_2'].str.split(' ',1,expand=True)
df_yoxi_1[['上車經度','上車緯度']] = df_yoxi_1['Why2'].str.split(' ',1,expand=True)
```

```python
df_yoxi_1.drop(['Why','Why2'],axis=1)
```

```python
df_yoxi_1[['Why','Why2']] = df_yoxi_1['下車經緯度_2'].str.split(' ',1,expand=True)
df_yoxi_1[['下車經度','下車緯度']] = df_yoxi_1['Why2'].str.split(' ',1,expand=True)
```

```python
df_yoxi_1 = df_yoxi_1.drop(['上車經緯度_1','下車經緯度_1', '下車經緯度_2', '上車經緯度_2'],axis=1)
```

```python
df_yoxi_1 = df_yoxi_1.drop(['Why','Why2'],axis=1)
df_yoxi_1
```

```python
df_yoxi_1.describe()
```

```python
df_yoxi_2 = df_yoxi_1
```

```python
df_yoxi_2.dropna(inplace=True)
```

```python
df_yoxi_2.drop_duplicates(inplace=True)
```

```python
df_yoxi_2
```

```python
kepler_map = KeplerGl(height=400)
```

```python
df_yoxi_3 = df_yoxi_2[['上車經度', '上車緯度', '下車經度', '下車緯度']].astype('float')
df_yoxi_3
```

```python
kepler_map.add_data(data=df_yoxi_3, name="Yoxi")
```

```python
kepler_map.save_to_html(file_name='Yoxi.html', 
                        data={"Yoxi": df_yoxi_3}
                        )
```
