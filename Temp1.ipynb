```python
df_iRent_GPS.columns = ['租車訂單號碼', 'GPS傳送時間', '緯度', '經度', '車輛行駛目前里程數', '車輛行駛時數']
```

```python
print(len(df_iRent_GPS['租車訂單號碼'].unique()))
```

```python
df_iRent_GPS.dropna(inplace=True)
df_iRent_GPS.drop_duplicates(inplace=True)
```

```python
df_iRent_GPS.describe()
```

```python
df_iRent_GPS_2 = df_iRent_GPS
```

```python
df_iRent_GPS_2['key'] = df_iRent_GPS_2['租車訂單號碼'].astype('str') + df_iRent_GPS_2['GPS傳送時間']
```

```python
df_iRent_GPS_2
```

```python
result_1 = df_iRent_GPS.groupby('租車訂單號碼').agg({'GPS傳送時間':'min'})
result_1 = result_1.reset_index()
result_1['key'] = result_1['租車訂單號碼'].astype('str') + result_1['GPS傳送時間']
```

```python
result_up = pd.merge(result_1, df_iRent_GPS_2, left_on="key", right_on="key")
result_up.rename(columns={'經度':'上車經度', '緯度':'上車緯度'}, inplace=True)
```

```python
result_up
```

```python
result_2 = df_iRent_GPS.groupby('租車訂單號碼').agg({'GPS傳送時間':'max'})
result_2 = result_2.reset_index()
result_2['key'] = result_2['租車訂單號碼'].astype('str') + result_2['GPS傳送時間']
```

```python
result_stop = pd.merge(result_2, df_iRent_GPS_2, left_on="key", right_on="key")
result_stop.rename(columns={'經度':'下車經度', '緯度':'下車緯度'}, inplace=True)
```

```python
result_stop.drop(['GPS傳送時間_x', 'key', '租車訂單號碼_y', 'GPS傳送時間_y','車輛行駛目前里程數','車輛行駛時數'],axis=1, inplace=True)
```

```python
result_up.rename(columns={'經度':'下車經度', '緯度':'下車緯度'}, inplace=True)
result_up.drop(['GPS傳送時間_x', 'key', '租車訂單號碼_y', 'GPS傳送時間_y','車輛行駛目前里程數','車輛行駛時數'],axis=1, inplace=True)
result_up
```

```python
result = pd.merge(result_up, result_stop, left_on="租車訂單號碼_x", right_on="租車訂單號碼_x")
```

```python
result = result[['上車經度', '上車緯度', '下車經度', '下車緯度']].astype('float')
#result = result[['租車訂單號碼_x','上車經度', '上車緯度', '下車經度', '下車緯度']].astype('float')
```

```python
result.dropna(inplace=True)
result.drop_duplicates(inplace=True)
```

```python
result.sort_values(by='下車經度')
result = result.drop(index=2973)
result
```

```python
kepler_map = KeplerGl(height=400)
```

```python
kepler_map.add_data(data=result, name="iRentGPS2")
```

```python
kepler_map.save_to_html(file_name='iRentGPS2.html', 
                        data={"iRentGPS2": result}
                        )
```
